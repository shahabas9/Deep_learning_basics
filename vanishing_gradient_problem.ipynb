```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
x,y = make_moons(n_samples=250,noise=0.05,random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=100)
plt.show()
```

```python
model=Sequential()
model.add(Dense(10,activation="sigmoid",input_dim=2))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
initial_weight=model.get_weights()[0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=36)
```

```python
model.fit(x_train,y_train,epochs=100)
```

Solution for vanishing gradient

```python
model=Sequential()
model.add(Dense(10,activation="sigmoid",input_dim=2))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=36)
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
# here it is using reducing the model complexity
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.astype('int'),clf=model,legend=2)
```

```python
# another method is using relu activation function
```

```python
model=Sequential()
model.add(Dense(10,activation="relu",input_dim=2))
model.add(Dense(10,activation="relu"))
model.add(Dense(10,activation="relu"))
model.add(Dense(10,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=36)
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.astype('int'),clf=model,legend=2)
```

```python
# next solution is proper weight initialization
```

```python
model=Sequential()
model.add(Dense(10,activation="relu",input_dim=2,kernel_initializer='he_normal'))
model.add(Dense(10,activation="relu",kernel_initializer='he_normal'))
model.add(Dense(10,activation="relu",kernel_initializer='he_normal'))
model.add(Dense(1,activation="sigmoid",kernel_initializer='he_normal'))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=36)
```

```python
model.fit(x_train,y_train,epochs=100)
```

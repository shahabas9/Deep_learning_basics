```python
from keras.models import Model
from keras.layers import *
```

```python

x=Input(shape=(3,))
```

```python
hidden_1=Dense(128,activation="relu")(x)
hidden_2=Dense(64,activation="relu")(hidden_1)
```

```python
output_1=Dense(1,activation="linear")(hidden_2)
output_2=Dense(1,activation="sigmoid")(hidden_2)
```

```python
model=Model(inputs=x,outputs=[output_1,output_2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```

```python
input_a=Input(shape=(32))
input_b=Input(shape=(128))
```

```python
# create a branch for input a

x1=Dense(8,activation="relu")(input_a)
x2=Dense(4,activation="relu")(x1)
```

```python
y1=Dense(64,activation="relu")(input_b)
y2=Dense(32,activation="relu")(y1)
y3=Dense(8,activation="relu")(y2)
```

```python
combined_value=concatenate([x2,y3])
```

```python
z=Dense(2,activation="relu")(combined_value)
z1=Dense(1,activation="linear")(z)
```

```python
model=Model(inputs=[input_a,input_b],outputs=z1)
```

```python
model.summary()
```

```python
plot_model(model,show_shapes=True)
```
